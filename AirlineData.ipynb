```python
#plotting values - No. of Flights v/s Airline

data['airline'].value_counts(ascending=True).plot.barh( color = ['red', 'blue'])
plot.title("Airlines with frequencies")
plot.xlabel(" Number of flights")
plot.ylabel(" Airlines")
plot.show()
```

```python
data['departure_time'].value_counts()
```

```python
#plotting values - Departure Time v/s No. of Flights

plot.bar( data['departure_time'].value_counts().index , data['departure_time'].value_counts().values, color = ['blue', 'green'] )
plot.title("Departure Time")
plot.xlabel("Departure Time")
plot.ylabel("Departure Freq")
plot.show()
```

```python
#plotting values - Arrival Time v/s No. of Flights

plot.bar( data['arrival_time'].value_counts().index, data['arrival_time'].value_counts().values, color = ['grey', 'black'])
plot.title("Arrival Time")
plot.xlabel("Arrival Time")
plot.ylabel("Arrival Freq")
plot.show()
```

```python
#plotting values - Source and Destination City v/s No. of Flights

plot.figure( figsize= (16,4))

plot.subplot(1,2,1)

plot.barh( data['source_city'].value_counts().index , data['source_city'].value_counts().values, color = ['yellow', 'green'])
plot.title("Source Cities with No. of flights")
plot.ylabel("Cities")
plot.xlabel("No. of flights")

plot.subplot(1,2,2)

plot.barh( data['destination_city'].value_counts().index , data['destination_city'].value_counts().values, color = ['brown', 'orange'])
plot.title("Destination Cities with No. of flights")
plot.ylabel("Cities")
plot.xlabel("No. of flights")

plot.show()
```
